### Importing the required libraries

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

### Importing the data

```python
df = pd.read_csv('owid-covid-data.csv')
```

#### Replacing the `nan` values 

```python
df = df.fillna(0)
df
```

### Taking all data from the last day of 2020 only and storing it into a new dataframe known as EOY_DF

```python
EOY_DF= df[(df['date']=='12/31/2020')] 
```

### Total cases VS Total Deaths
How many cases there are in "x" country and how many deaths they have for their entire cases
E.G : They had ~1000 people who have been diagnosed and had 10 people who died who had COVID-19,
what is the percentage of people who died who had COVID-19 <br>

Firstly I'll explore the percentage of deaths for cases `worldwide` <br>
Secondly I'll explore the percentage of deaths for cases of each country

##### Percentage of total deaths out of total cases worldwide 

```python
cases_deaths_df = EOY_DF[['location','total_cases','total_deaths']]
```

```python
world_deaths_percentage =round((int(cases_deaths_df[cases_deaths_df['location']=='World']['total_deaths'])/int(cases_deaths_df[cases_deaths_df['location']=='World']['total_cases']))*100,2)
```

### The percentage of people who have died of COVID-19 out of the total cases on 2020-12-31(final day of 2020) :

```python
world_deaths_percentage
```

### Percentage of total deaths out of total cases for each country 


```python
cases_deaths_df = EOY_DF[(EOY_DF['location']!='Africa')&(EOY_DF['location']!='Asia')&(EOY_DF['location']!='Europe')&(EOY_DF['location']!='European Union')&(EOY_DF['location']!='International')&(EOY_DF['location']!='North America')&(EOY_DF['location']!='Oceania')&(EOY_DF['location']!='South America')&(EOY_DF['location']!='World')]
```

```python
#removing data from locations that are not countries (such as continents and worldwide)
cases_deaths_df = cases_deaths_df[['location','total_deaths','total_cases']]
```

```python
#checking for any np.nan values 
0.0 in cases_deaths_df['total_deaths'].tolist()
```

```python
#print(0) if((row['total_cases']==0)|(row['total_deaths']==0)) else
location = []
deaths_percentage = []
for i, row in cases_deaths_df.iterrows():
    if((row['total_cases']==0)&(row['total_deaths']==0)): 
        deaths_percentage.append(0)
    else:
        location.append(row['location'])
        deaths_percentage.append(round((row['total_deaths']/row['total_cases'])*100,2))
    
```

With the data successfully made I now want to be able to put this into a dataframe so that I can display this in a more appealing format

```python
deaths_percentage_df = pd.DataFrame(list(zip(location,deaths_percentage)),columns=['location','deaths_percentage'])
```

```python
deaths_percentage_df
```

### Show what percentage of the population got COVID-19 

```python
cases_population_df = EOY_DF[(EOY_DF['location']!='Africa')&(EOY_DF['location']!='Asia')&(EOY_DF['location']!='Europe')&(EOY_DF['location']!='European Union')&(EOY_DF['location']!='International')&(EOY_DF['location']!='North America')&(EOY_DF['location']!='Oceania')&(EOY_DF['location']!='South America')&(EOY_DF['location']!='World')]
cases_population_df = cases_population_df[['location','total_cases','population']] #taking only columns: location,total_cases,population
cases_population_df
```

### What country is the highest infection rate compared to the population

```python
cases_percentage = []
for i, row in cases_population_df.iterrows():
    cases_percentage.append(round((row['total_cases']/row['population'])*100,2))
cases_percentage_df = pd.DataFrame(list(zip(location,cases_percentage)),columns=['location','Cases_percentage'])
```

#### Country with the highest infection percentage

```python
cases_percentage_df[cases_percentage_df['Cases_percentage']==cases_percentage_df['Cases_percentage'].max()]
```

### Countries with the highest death per population 

```python
# Top 10 countries with the highest deaths per population
deaths_percentage_df.sort_values(by=['deaths_percentage'],ascending=False).head(10)
```

### Breaking things down by continent

Dataframe will be made for the following : <br>
[x] 1. Percentages of deaths by cases  
[x] 2. Percentage of cases by population 

```python
#Filtering data
continents = EOY_DF[(EOY_DF['location']=='Africa')|(EOY_DF['location']=='Asia')|(EOY_DF['location']=='Europe')|(EOY_DF['location']=='North America')|(EOY_DF['location']=='South America')]
EOY_DF_Continents = continents[['location','total_cases','total_deaths','population']]
#Collecting all the continents and storing then into a list
conts = EOY_DF_Continents['location'].tolist()
#Initializing a list containing the percentage of the continents total deaths by the total cases
cont_deaths = []
for i, row in EOY_DF_Continents.iterrows():
    cont_deaths.append(round((row['total_deaths']/row['total_cases'])*100,2))
#Initializing a list containing the percentage of the continents total cases by the population
cont_cases = []
for i, row in EOY_DF_Continents.iterrows():
    cont_cases.append(round((row['total_cases']/row['population'])*100,2))
#storing the list values into a list so that it is easier to create a dataframe 
continent_dict = {
    'continent':conts,
    'Cases by total population (%) ':cont_cases,
    'Deaths by total cases (%)':cont_deaths
}
continent_df = pd.DataFrame.from_dict(continent_dict,orient='index').transpose()
```

#### Percentage of vaccincations by population (continent)

```python
'''
This will store all the percentages into a list which will be in the same order of the continents 
'''
percentages = []
for i, row in continents.iterrows():
    percentages.append(round((row['total_vaccinations']/row['population'])*100,2))
continent_vax_dict= {
    'continent': continents['location'].tolist(),
    'percentage vaccinated':percentages
}
continent_vax_df = pd.DataFrame.from_dict(continent_vax_dict,orient='index').transpose()
```

## Data Visualization

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.set_title('Continent Percentage of Vaccinations by their populations')
ax.bar(continent_vax_df['continent'].tolist(),continent_vax_df['percentage vaccinated'].tolist())
ax.grid(axis='y')
ax.set_ylabel("Percentage")
ax.set_xlabel("Continents")
```

```python
top_10_countries_vax_percentage= pop_vax_df.sort_values(by=['vax_percentage'],ascending=False).head(10)
fig = plt.figure()
ax = fig.add_axes([0,0,2,2.5])
ax.set_title("Percentage of the top 10 countries vaccinations by population ")
ax.bar(top_10_countries_vax_percentage['country'].tolist(),top_10_countries_vax_percentage['vax_percentage'].tolist())
ax.grid(axis='y')
ax.set_ylabel('Percentage')
ax.set_xlabel('Country')
for index,data in enumerate(top_10_countries_vax_percentage['vax_percentage'].tolist()):
    plt.text(x = index-0.3, y= data + 0.2,s=f"{data}%", fontdict=dict(fontsize=15))
```
